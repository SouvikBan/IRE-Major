```
from googletrans import Translator
```

```
translator = Translator()
data = ['Tamil Nadu', 'Maharashtra', 'Trivandrum', 'Salem', 'Trichy', 'Coimbatore', 'Pondicherry']
for place in data:
    text = translator.translate(place, src='en', dest='hi')
    print(text.text, end='  ')
```
